```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../python/')
import angles
import instrument_mm
```

# Making Plots for HWP

```python
# Define range of HWP angles
delta_HWP = 0.5 # retardance of the waveplate in waves
angles = np.linspace(0, 90, 200)
m21_vals, m31_vals, m41_vals = [], [], []

# Compute Mueller matrix for each angle
for angle in angles:
    mm = instrument_mm.MODHIS_no_wollaston_or_rotations(delta_HWP=delta_HWP, HWP_ang=angle)
    mm_abs = np.abs(mm)  # Ensure absolute values are used for plotting
    m21_vals.append(mm_abs[1, 1] * 100)  # 2nd row, 2nd column (Q Efficiency)
    m31_vals.append(mm_abs[2, 1] * 100)  # 3rd row, 2nd column (U Efficiency)
    m41_vals.append(mm_abs[3, 1] * 100)  # 4th row, 2nd column (V Efficiency)

# Plot Q Efficiency
fig1, ax1 = plt.subplots(figsize=(8, 4))
ax1.plot(angles, m21_vals, label='Q Efficiency')
ax1.axvline(0, color='red', linestyle='--', linewidth=1)
ax1.axvline(45, color='red', linestyle='--', linewidth=1)
ax1.set_xlabel("HWP Angle ()")
ax1.set_ylabel("Polarimetric Efficiency (%)")
ax1.set_title("Stokes Q Modulation")
# ax1.legend()
plt.tight_layout()
plt.show()

# Plot U Efficiency
fig2, ax2 = plt.subplots(figsize=(8, 4))
ax2.plot(angles, m31_vals, label='U Efficiency')
ax2.axvline(22.5, color='red', linestyle='--', linewidth=1)
ax2.axvline(67.5, color='red', linestyle='--', linewidth=1)
ax2.set_xlabel("HWP Angle ()")
ax2.set_ylabel("Polarimetric Efficiency (%)")
ax2.set_title("Stokes U Modulation")
# ax2.legend()
plt.tight_layout()
plt.show()

# Plot V Efficiency
fig3, ax3 = plt.subplots(figsize=(8, 4))
ax3.plot(angles, m41_vals, label='V Efficiency')
ax3.set_xlabel("HWP Angle ()")
ax3.set_ylabel("Polarimetric Efficiency (%)")
ax3.set_title("Stokes V Modulation")
ax3.legend()
plt.tight_layout()
plt.show()
```
