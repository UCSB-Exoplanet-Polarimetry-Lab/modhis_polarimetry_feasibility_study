# Making Plots for HWP

```python
import numpy as np
import matplotlib.pyplot as plt
import instrument_mm  # ensure this is your module name

# Define range of HWP angles
delta_HWP = 0.5  # retardance of the waveplate in waves
angles = np.linspace(0, 180, 181)
m21_vals, m31_vals, m41_vals = [], [], []

# Compute Mueller matrix elements in 2nd column (excluding first row)
for angle in angles:
    mm = instrument_mm.MODHIS_no_wollaston_or_rotations(delta_HWP=delta_HWP, HWP_ang=angle)
    m21_vals.append(mm[1, 1])  # 2nd row, 2nd column (column 1)
    m31_vals.append(mm[1, 2])  # 3rd row, 2nd column
    m41_vals.append(mm[1, 3])  # 4th row, 2nd column

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].plot(angles, m21_vals)
axs[0].set_title("Q Modulation for HWP Angle")
axs[0].set_xlabel("WP Angle (degrees)")
axs[0].set_ylabel("M[1,1]")

axs[1].plot(angles, m31_vals)
axs[1].set_title("U Modulation for HWP Angle")
axs[1].set_xlabel("WP Angle (degrees)")
axs[1].set_ylabel("M[1,2]")

axs[2].plot(angles, m41_vals)
axs[2].set_title("V Modulation for HWP Angle")
axs[2].set_xlabel("WP Angle (degrees)")
axs[2].set_ylabel("M[1,3]")

plt.tight_layout()
plt.show()

# Print min and max values
for vals, label in zip([m21_vals, m31_vals, m41_vals], ['M[1,1]', 'M[2,1]', 'M[3,1]']):
    min_val = np.min(vals)
    max_val = np.max(vals)
    min_angle = angles[np.argmin(vals)]
    max_angle = angles[np.argmax(vals)]
    efficiency_percentage = ((max_val - min_val) / 2) * 100
    print(f"{label} - Min: {min_val:.5f} at {min_angle:.1f}, Max: {max_val:.5f} at {max_angle:.1f}, Efficiency %: {efficiency_percentage:.5f}")
```
