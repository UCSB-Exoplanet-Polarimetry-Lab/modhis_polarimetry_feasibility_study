# Making Plots for HWP

```python
import numpy as np
import matplotlib.pyplot as plt
import instrument_mm  # ensure this module is correctly imported

# Define range of HWP angles
delta_HWP = 0.5  # retardance of the waveplate in waves
angles = np.linspace(0, 180, 181)
row2_sums, row3_sums, row4_sums = [], [], []

# Compute summed rows for each angle
for angle in angles:
    mm = instrument_mm.MODHIS_no_wollaston_or_rotations(delta_HWP=delta_HWP, HWP_ang=angle)
    row2_sums.append(np.sum(mm[1, :]))  # 2nd row
    row3_sums.append(np.sum(mm[2, :]))  # 3rd row
    row4_sums.append(np.sum(mm[3, :]))  # 4th row

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].plot(angles, row2_sums)
axs[0].set_title("Sum of 2nd Row vs HWP Angle")
axs[0].set_xlabel("HWP Angle (degrees)")
axs[0].set_ylabel("Sum of Row 2")

axs[1].plot(angles, row3_sums)
axs[1].set_title("Sum of 3rd Row vs HWP Angle")
axs[1].set_xlabel("HWP Angle (degrees)")
axs[1].set_ylabel("Sum of Row 3")

axs[2].plot(angles, row4_sums)
axs[2].set_title("Sum of 4th Row vs HWP Angle")
axs[2].set_xlabel("HWP Angle (degrees)")
axs[2].set_ylabel("Sum of Row 4")

plt.tight_layout()
plt.show()

# Print min and max values
for vals, label in zip([row2_sums, row3_sums, row4_sums], ['Row 2 Sum', 'Row 3 Sum', 'Row 4 Sum']):
    min_val = np.min(vals)
    max_val = np.max(vals)
    min_angle = angles[np.argmin(vals)]
    max_angle = angles[np.argmax(vals)]
    efficiency_percentage = ((max_val - min_val) / 2) * 100
    print(f"{label} - Min: {min_val:.5f} at {min_angle:.1f}, Max: {max_val:.5f} at {max_angle:.1f}, Efficiency %: {efficiency_percentage:.5f}")
```
