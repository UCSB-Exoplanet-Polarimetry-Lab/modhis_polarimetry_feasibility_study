```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../python/')
import angles
import instrument_mm
```

# Making Plots for HWP

```python
# Define range of HWP angles
delta_HWP = 0.5 # retardance of the waveplate in waves
angles = np.linspace(0, 180, 181)
m22_vals, m33_vals, m44_vals = [], [], []

# Compute Mueller matrix for each angle
for angle in angles:
    mm = instrument_mm.MODHIS_no_wollaston_or_rotations(delta_HWP = delta_HWP, 
        HWP_ang = angle)
    m22_vals.append(mm[1, 1])
    m33_vals.append(mm[2, 2])
    m44_vals.append(mm[3, 3])

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].plot(angles, m22_vals)
axs[0].set_title("M[1,1] vs HWP Angle")
axs[0].set_xlabel("HWP Angle (degrees)")
axs[0].set_ylabel("M[1,1]")

axs[1].plot(angles, m33_vals)
axs[1].set_title("M[2,2] vs HWP Angle")
axs[1].set_xlabel("HWP Angle (degrees)")
axs[1].set_ylabel("M[2,2]")

axs[2].plot(angles, m44_vals)
axs[2].set_title("M[3,3] vs HWP Angle")
axs[2].set_xlabel("HWP Angle (degrees)")
axs[2].set_ylabel("M[3,3]")

plt.tight_layout()
plt.show()

# Print min and max values
for idx, (vals, label) in enumerate(zip([m22_vals, m33_vals, m44_vals], ['M[1,1]', 'M[2,2]', 'M[3,3]'])):
    min_val = np.min(vals)
    max_val = np.max(vals)
    min_angle = angles[np.argmin(vals)]
    max_angle = angles[np.argmax(vals)]
    # Print min/max values/angles, and efficiency %
    efficiency_percentage = ((max_val - min_val) / 2) * 100
    print(f"{label} - Min: {min_val:.5f} at {min_angle:.1f}, Max: {max_val:.5f} at {max_angle:.1f}, Efficiency %: {efficiency_percentage:.5f}")
```
